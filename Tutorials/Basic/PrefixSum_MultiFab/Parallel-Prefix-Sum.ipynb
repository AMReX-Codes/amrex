```python
import yt
import numpy as np
```

```python
filename = "plt00001"
```

```python
ds = yt.load(filename)
```

```python
cg = ds.covering_grid(left_edge=ds.domain_left_edge, dims=ds.domain_dimensions, level=0)
coords = cg.fcoords

xs = coords[:,0]

dx = np.max(xs[1:] - xs[:-1])

phi_prefix_sum = cg["phi_psum"][:,:,:].d
```

```python
for i in range(ds.domain_dimensions[0]):
    for j in range(ds.domain_dimensions[1]):
        for k in range(ds.domain_dimensions[2]):
            print("({},{},{}): ".format(i,j,k) + str(phi_prefix_sum[i,j,k]))
```
